```python
# SimpleAgent: No memory, treats each question independently
from stocks_agent.simple_agent import SimpleAgent

simple = SimpleAgent(model="gpt-4o-mini")
print("✓ SimpleAgent initialized")
```

```python
# Test basic query
response = await simple.ask("What's AAPL's PE ratio?")
print(response)
```

```python
# Test another query (no memory - won't remember AAPL)
response = await simple.ask("What about the earnings?")
print(response)  # Won't know you meant AAPL
```

```python

# Test another query - more extended question
response = await simple.ask("Recent news and sentiment on AAPL? EPS trend and analysts change in sentiment")
print(response)  
```

```python
# ConversationAgent: Remembers context, auto-tracks tickers, handles follow-ups
  # Cell 1: Import and initialize ConversationAgent
from stocks_agent.conversation_agent import ConversationAgent

conv = ConversationAgent(track_tickers=True, model="gpt-4o-mini")
print("✓ ConversationAgent initialized")
```

```python
# Conv agent: Ask the same question about AAPL

response = await conv.ask("Recent news and sentiment on AAPL? EPS trend and analysts change in sentiment")
print(response)
```

```python
response = await conv.ask("Follow-up: what are the competitors trends?")
print(response)
```

```python
# get the internal context:
tickers = conv.get_tickers()
print(f"Tickers discussed: {tickers}")
print(f"Current ticker: {conv.current_ticker}")
```
